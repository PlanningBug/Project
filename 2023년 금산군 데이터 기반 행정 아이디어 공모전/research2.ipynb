## 휴게음식점 현황

```python
import pandas as pd

df = pd.read_csv('/Users/raekkkky/Desktop/충청남도_금산군_휴게음식점_20220331.csv', encoding='euc-kr')
df.head()
```

```python
def check_contains(text):
    keywords = ['금산읍', '금성면', '제원면', '부리면', '군북면', '남일면', '남이면', '진산면', '복수면', '추부면']
    for keyword in keywords:
        if keyword in text:
            return keyword
    return None

# '소재지(도로명)' 열에 대해 check_contains 함수를 적용하여 '단어' 열을 생성
df['단어'] = df['소재지(도로명)'].apply(check_contains)
df = df[['단어']]

result = df['단어'].value_counts().reset_index()
result.columns = ['단어', '합산']

# 원하는 순서대로 정렬하기 위한 순서 리스트
desired_order = ['군북면', '금산읍', '금성면', '남이면', '남일면', '복수면', '부리면', '제원면', '진산면', '추부면']

# 순서 리스트를 기반으로 데이터프레임을 정렬
result['단어'] = pd.Categorical(result['단어'], categories=desired_order, ordered=True)
result = result.sort_values(by='단어')

# 결과 출력
print(result)
```

```python
result = result['합산']
print(result)
```

## 유해야생동물

```python
df5 = pd.read_csv('/Users/raekkkky/Desktop/충청남도_금산군 유해야생동물 포획데이터_20220630.csv')
df5.head()
```

```python
df5.info()
```

```python
is_na = df5.isna()

# 결측값 개수와 결측값이 아닌 값 개수 계산
na_counts = is_na.sum()  # 결측값 개수
not_na_counts = (~is_na).sum()  # 결측값이 아닌 값 개수

na_counts
```

```python
df5['포획일자'] = pd.to_datetime(df5['포획일자'], errors='coerce')

# 2021년 데이터 필터링
df5 = df5[(df5['포획일자'] >= '2021-01-01') & (df5['포획일자'] < '2022-01-01')]

# 결과 확인
df5.head()
```

```python
df5.tail()
```

```python
df5.info()
```

```python
def check_contains(text):
    keywords = ['금산읍', '금성면', '제원면', '부리면', '군북면', '남일면', '남이면', '진산면', '복수면', '추부면']
    for keyword in keywords:
        if keyword in text:
            return keyword
    return None

# '소재지(도로명)' 열에 대해 check_contains 함수를 적용하여 '단어' 열을 생성
df5['단어'] = df5['포획지상세주소'].apply(check_contains)

# '단어' 열에 따른 포획개체수 합을 계산
result = df5.groupby('단어')['포획개체수'].sum().reset_index()

# 원하는 순서대로 정렬하기 위한 순서 리스트
desired_order = ['군북면', '금산읍', '금성면', '남이면', '남일면', '복수면', '부리면', '제원면', '진산면', '추부면']

# 순서 리스트를 기반으로 데이터프레임을 정렬
result['단어'] = pd.Categorical(result['단어'], categories=desired_order, ordered=True)
result = result.sort_values(by='단어')

# 결과 출력
print(result)
```

```python
total_sum = result['포획개체수'].sum()

# 결과 출력
print("총 합:", total_sum)
```
