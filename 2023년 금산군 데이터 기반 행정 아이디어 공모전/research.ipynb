# 음식물 쓰레기

```python
import pandas as pd

# CSV 파일을 데이터프레임으로 읽기
df = pd.read_csv('/Users/raekkkky/Desktop/충청남도_금산군_음식물쓰레기 다량배출사업장현황_20210208.csv', encoding='EUC-KR')
df
```

```python
df = df[['소재지 도로명주소', '소재지 지번주소']]
df
```

```python
df.columns = [1, 2]
df
```

```python
import numpy as np

df[3] = np.where(df[1].isna(), df[2], df[1])
df
```

```python
df = df[[3]]
df
```

```python
# '금산읍', '금성면', '제원면', '부리면', '군북면', '남일면', '남이면', '진산면', '복수면', '추부면' 중 하나가 포함되어 있는지 확인하는 함수
def check_contains(text):
    keywords = ['금산읍', '금성면', '제원면', '부리면', '군북면', '남일면', '남이면', '진산면', '복수면', '추부면']
    for keyword in keywords:
        if keyword in text:
            return keyword
    return None

# '주소' 열에서 check_contains 함수를 적용하여 특정 단어를 확인하고, 해당 단어를 '단어' 열에 추가
df['단어'] = df[3].apply(check_contains)

df
```

```python
result = df['단어'].value_counts().reset_index()
result.columns = ['단어', '합산']

print(result)
```
