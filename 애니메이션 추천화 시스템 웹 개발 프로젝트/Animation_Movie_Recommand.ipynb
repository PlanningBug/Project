## 2. ì½˜í…ì¸  ê¸°ë°˜ ì¶”ì²œí™” ì•Œê³ ë¦¬ì¦˜ ì ìš© [ì¥ë¥´ ê¸°ë°˜ + ì‚¬ìš©ì ì˜í™” 1ê°œ ì…ë ¥ + ê°€ì¤‘í‰ì  ì‚¬ìš© + top10ê°œ ì¶”ì²œ]

```python
import pandas as pd
pd.set_option('display.max_columns', None)

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° ë° ì „ì²˜ë¦¬
data_path = "/Users/raekkkky/Desktop/DNA/[24 1í•™ê¸°] ğŸ†ìŠ¹ìŠ¹ì¥êµ¬/9ì£¼ì°¨/Animation_Movies.csv"
df = pd.read_csv(data_path)
columns = [
    'ì•„ì´ë””', 'ì œëª©', 'í‰ê·  íˆ¬í‘œ', 'íˆ¬í‘œ ìˆ˜', 'ìƒíƒœ', 'ê°œë´‰ì¼',
    'ìˆ˜ìµ', 'ëŸ¬ë‹íƒ€ì„', 'ì„±ì¸ ì—¬ë¶€', 'ë°°ê²½ ì´ë¯¸ì§€ ê²½ë¡œ', 'ì˜ˆì‚°', 'í™ˆí˜ì´ì§€',
    'IMDB ì•„ì´ë””', 'ì›ë˜ ì–¸ì–´', 'ì›ë˜ ì œëª©', 'ë¦¬ë·°',
    'ì¸ê¸°ë„', 'í¬ìŠ¤í„° ê²½ë¡œ', 'íƒœê·¸ë¼ì¸', 'ì¥ë¥´',
    'ì œì‘ íšŒì‚¬', 'ì œì‘ êµ­ê°€', 'ì‚¬ìš© ì–¸ì–´'
]
df.columns = columns
df = df.dropna(subset=['ì œëª©', 'ì›ë˜ ì œëª©'])
df = df.drop(columns=['ìƒíƒœ', 'ê°œë´‰ì¼', 'ë°°ê²½ ì´ë¯¸ì§€ ê²½ë¡œ', 'í™ˆí˜ì´ì§€', 'IMDB ì•„ì´ë””', 'í¬ìŠ¤í„° ê²½ë¡œ', 'íƒœê·¸ë¼ì¸', 'ì œì‘ íšŒì‚¬', 'ì œì‘ êµ­ê°€', 'ì‚¬ìš© ì–¸ì–´'])

# ì¶”ì²œì— ì‚¬ìš©í•  ë°ì´í„°í”„ë ˆì„ ìƒì„±
recommendation_df = df[['ì œëª©', 'ì¥ë¥´', 'í‰ê·  íˆ¬í‘œ', 'íˆ¬í‘œ ìˆ˜']].copy()
```

```python
# ê°€ì¤‘ í‰ì  ê³„ì‚° í•¨ìˆ˜
def weighted_rating(x, m, C):
    v = x['íˆ¬í‘œ ìˆ˜']
    R = x['í‰ê·  íˆ¬í‘œ']
    return (v / (v + m) * R) + (m / (m + v) * C)

# ì „ì²´ ë°ì´í„°ì˜ í‰ê·  í‰ì ê³¼ ìµœì†Œ íˆ¬í‘œ ìˆ˜ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì„¤ì •
C = recommendation_df['í‰ê·  íˆ¬í‘œ'].mean()
m = recommendation_df['íˆ¬í‘œ ìˆ˜'].quantile(0.90)
```

```python
# ì¥ë¥´ ê¸°ë°˜ ì½˜í…ì¸  í•„í„°ë§ ì¶”ì²œ ì‹œìŠ¤í…œ êµ¬í˜„ (ê°€ì¤‘í‰ì  ë°˜ì˜)
def genre_recommendation(title, df, top_n=10):
    movie_idx = df[df['ì œëª©'] == title].index[0]
    movie_genres = df.at[movie_idx, 'ì¥ë¥´'].split(", ")

    scores = []
    for index, row in df.iterrows():
        current_genres = row['ì¥ë¥´'].split(", ")
        common_genres = set(movie_genres) & set(current_genres)
        score = len(common_genres) / len(set(movie_genres))

        if row['íˆ¬í‘œ ìˆ˜'] >= m:
            weighted_score = score * weighted_rating(row, m, C)
        else:
            weighted_score = score * row['í‰ê·  íˆ¬í‘œ']

        scores.append((index, weighted_score))

    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    recommended_movies = [df.iloc[idx]['ì œëª©'] for idx, score in scores[1:top_n+1]]

    return recommended_movies
```

```python
# ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
user_input = input("ì ~ ë‹¹ì‹ ì´ ë³¸ ì˜í™”ì˜ ì œëª©ì„ ì…ë ¥í•´ ë³´ì‹œì˜·! : ")
recommended_movies = genre_recommendation(user_input, recommendation_df)
```

```python
# ê²°ê³¼ ì¶œë ¥
print("ë‹¹ì‹ ì´ ë³¸ ì˜í™”ì™€ ë¹„ìŠ·í•œ ì¥ë¥´ì˜ ì˜í™” ì¶”ì²œ ëª©ë¡[top10!!]ì…ë‹ˆë‹¤:")
for movie in recommended_movies:
    print(movie)
```
