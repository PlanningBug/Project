## 2. 콘텐츠 기반 추천화 알고리즘 적용 [장르 기반 + 사용자 영화 1개 입력 + 가중평점 사용 + top10개 추천]

```python
import pandas as pd
pd.set_option('display.max_columns', None)

# 데이터 불러오기 및 전처리
data_path = "/Users/raekkkky/Desktop/DNA/[24 1학기] 🏆승승장구/9주차/Animation_Movies.csv"
df = pd.read_csv(data_path)
columns = [
    '아이디', '제목', '평균 투표', '투표 수', '상태', '개봉일',
    '수익', '러닝타임', '성인 여부', '배경 이미지 경로', '예산', '홈페이지',
    'IMDB 아이디', '원래 언어', '원래 제목', '리뷰',
    '인기도', '포스터 경로', '태그라인', '장르',
    '제작 회사', '제작 국가', '사용 언어'
]
df.columns = columns
df = df.dropna(subset=['제목', '원래 제목'])
df = df.drop(columns=['상태', '개봉일', '배경 이미지 경로', '홈페이지', 'IMDB 아이디', '포스터 경로', '태그라인', '제작 회사', '제작 국가', '사용 언어'])

# 추천에 사용할 데이터프레임 생성
recommendation_df = df[['제목', '장르', '평균 투표', '투표 수']].copy()
```

```python
# 가중 평점 계산 함수
def weighted_rating(x, m, C):
    v = x['투표 수']
    R = x['평균 투표']
    return (v / (v + m) * R) + (m / (m + v) * C)

# 전체 데이터의 평균 평점과 최소 투표 수를 기준으로 설정
C = recommendation_df['평균 투표'].mean()
m = recommendation_df['투표 수'].quantile(0.90)
```

```python
# 장르 기반 콘텐츠 필터링 추천 시스템 구현 (가중평점 반영)
def genre_recommendation(title, df, top_n=10):
    movie_idx = df[df['제목'] == title].index[0]
    movie_genres = df.at[movie_idx, '장르'].split(", ")

    scores = []
    for index, row in df.iterrows():
        current_genres = row['장르'].split(", ")
        common_genres = set(movie_genres) & set(current_genres)
        score = len(common_genres) / len(set(movie_genres))

        if row['투표 수'] >= m:
            weighted_score = score * weighted_rating(row, m, C)
        else:
            weighted_score = score * row['평균 투표']

        scores.append((index, weighted_score))

    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    recommended_movies = [df.iloc[idx]['제목'] for idx, score in scores[1:top_n+1]]

    return recommended_movies
```

```python
# 사용자 입력 받기
user_input = input("자 ~ 당신이 본 영화의 제목을 입력해 보시옷! : ")
recommended_movies = genre_recommendation(user_input, recommendation_df)
```

```python
# 결과 출력
print("당신이 본 영화와 비슷한 장르의 영화 추천 목록[top10!!]입니다:")
for movie in recommended_movies:
    print(movie)
```
