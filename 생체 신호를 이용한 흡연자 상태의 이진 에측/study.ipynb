```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 스타일 설정
sns.set(style="whitegrid")

# train_df의 모든 열에 대한 boxplot 그리기
plt.figure(figsize=(16, 8))
sns.boxplot(data=train_df, palette='viridis')
plt.title('Boxplot of Data Distribution for Each Column')
plt.xticks(rotation=45, ha='right')  # X 축 레이블 회전
plt.show()
```

```python
# train_df의 이상치를 확인하고 제거할 열 선택
columns_to_check = ['age', 'height(cm)', 'weight(kg)', 'waist(cm)', 'eyesight(left)', 'eyesight(right)', 'hearing(left)', 'hearing(right)',
                    'systolic', 'relaxation', 'fasting blood sugar', 'Cholesterol', 'triglyceride', 'HDL', 'LDL', 'hemoglobin', 'serum creatinine',
                    'AST', 'ALT', 'Gtp']

# 이상치 제거 함수 정의
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# 이상치 제거 적용
train_df_no_outliers = remove_outliers(train_df, columns_to_check)

# 이상치 제거 후의 boxplot
plt.figure(figsize=(16, 8))
sns.boxplot(data=train_df_no_outliers[columns_to_check], palette='viridis')
plt.title('Boxplot of Data Distribution for Each Column (No Outliers)')
plt.xticks(rotation=45, ha='right')  # X 축 레이블 회전
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 시각화 스타일 설정
sns.set(style="whitegrid")

# 선택한 칼럼들
selected_columns_with_outliers = ["eyesight(left)", "eyesight(right)", "fasting blood sugar", "triglyceride",
                                   "serum creatinine", "LDL", "hemoglobin", "AST", "ALT", "Gtp"]

selected_columns_no_outliers = [col for col in train_df.columns if col not in ["age", "age_group"] + selected_columns_with_outliers]

# 이상치 제거 함수 정의
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# 이상치 제거 적용 (10개의 칼럼)
train_df_no_outliers_with = remove_outliers(train_df, selected_columns_with_outliers)

# 나이대(age_group)별로 선택한 칼럼의 분포를 보여주는 바이올린플롯과 막대그래프 (이상치 제거 전)
plt.figure(figsize=(24, 12))  # subplot의 크기를 조절

# 총 22개의 subplot을 4행 6열로 배치
for i, column in enumerate(selected_columns_with_outliers + selected_columns_no_outliers):
    plt.subplot(4, 6, i+1)
    
    # 막대그래프
    if column in ["hearing(left)", "hearing(right)", "Urine protein", "dental caries", "smoking"]:
        sns.countplot(x=column, hue='age_group', data=train_df, palette='viridis')
    else:
        # 바이올린플롯
        sns.violinplot(x='age_group', y=column, data=train_df, palette='viridis')
    
    plt.title(f'Distribution of {column} by Age Group (Original Data)')
    plt.xlabel('Age Group')
    plt.ylabel(column)

plt.tight_layout()
plt.show()

# 나이대(age_group)별로 선택한 칼럼의 분포를 보여주는 바이올린플롯과 막대그래프 (이상치 제거 후)
plt.figure(figsize=(24, 12))  # subplot의 크기를 조절

# 총 22개의 subplot을 4행 6열로 배치
for i, column in enumerate(selected_columns_with_outliers + selected_columns_no_outliers):
    plt.subplot(4, 6, i+1)
    
    # 막대그래프
    if column in ["hearing(left)", "hearing(right)", "Urine protein", "dental caries", "smoking"]:
        sns.countplot(x=column, hue='age_group', data=train_df_no_outliers_with, palette='viridis')
    else:
        # 바이올린플롯
        sns.violinplot(x='age_group', y=column, data=train_df_no_outliers_with, palette='viridis')
    
    plt.title(f'Distribution of {column} by Age Group (No Outliers)')
    plt.xlabel('Age Group')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```
